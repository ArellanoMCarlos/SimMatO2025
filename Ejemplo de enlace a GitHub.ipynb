# Enlace a Github desde Google Colab
Antes de clonar el repositorio de GitHub, debemos dar acceso a los archivos del drive:

* En la pestaña lateral izquierda, hacer clic en el simbolo de la carpeta con nombre 'Archivos'
* Se abrirá un panel. En él, hacemos clic en el simbolo de carpeta con ícono del Drive. Esta carpeta lleva por nombre 'Montar Drive'
* Seleccionamos la cuenta con la que queremos que se sincronice y aceptamos el acceso.

Este proceso solo es necesario hacerlo una vez.

**Clonamos el repositorio de GitHub**

Vea que debemos sustituir `direccion_donde_guardar_el_repositorio` por la dirección del Drive donde queremos que se clone el repositorio. También, sustituímos `Usuario` por el usuario de GitHub y `Repositorio` por el nombre que le dimos al repositorio.

````python
%%bash

cd "/content/drive/MyDrive/direccion_donde_guardar_el_repositorio"
git clone https://github_pat_token_personal@github.com/Usuario/Repositorio.git
````

**Ingresamos nuestro Usuario y correo asociados a GitHGub**

````python
%%bash

cd "/content/drive/MyDrive/direccion_donde_guardar_el_repositorio/Repositorio"
git config --global user.email "correo@iteso.mx"
git config --global user.name "Usuario"
````

**Ingresamos nuestras credenciales para iniciar sesión en GitHub desde Google Colab**

Para generar el _token_ ingresamos a GitHub, hacemos clic en nuestra imágen de usuario y seleccionamos `Settings`, en el panel izquierdo seleccionamos `Developer settings` y finalmente hacemos clic en `Personal access tokens`. Podemos generar cualquier tipo de _token_ solo nos debemos asegurar de tener permisos de lectura y escritura.

La cadena de texto que nos arroje la sustituimos por los caracteres `github_pat_personaltoken` del enlace que da acceso al repositorio en la siguiente celda de código.

````python
%%bash

cd "/content/drive/MyDrive/direccion_donde_guardar_el_repositorio/Repositorio"
git remote set-url origin https://github_pat_personaltoken@github.com/Usuario/Repositorio.git
git remote -v
````

Al hacer una modificación en el repositorio, agregamos todos los archivos en la proxima modificación:
```
git add .
```
o  podemos especificar solo los archivos a actualizar:
```
git add nombre_del_archivo
```

podemos especificar un `commit` y cargamos los archivos o modificaciones a GitHub usando `git push`

````python
%%bash

cd "/content/drive/MyDrive/direccion_donde_guardar_el_repositorio/Repositorio"
git add .
git commit -m "Comentario del commit"
git push
````

De manera recursiva nos preguntará quién hace las modificaciones al repositorio, por lo que el siguiente código es útil para ejecutar solo una celda una vez clonado el repositorio y almacenado la url con el _token_ personal

````python
%%bash

cd "/content/drive/MyDrive/direccion_donde_guardar_el_repositorio/Repositorio"
git config --global user.email "correo@iteso.mx"
git config --global user.name "Usuario"
git add .
git commit -m "Comentario del commit"
git push
````

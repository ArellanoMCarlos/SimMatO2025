### SymPy

<img style="float: right; margin: 0px 0px 15px 15px;" src="http://www.sympy.org/static/images/logo.png" width="200px" height="100px" />

`SymPy` (Symbolic Python) es una librería de `Python` para matemáticas simbólicas. Su objetivo es convertirse en un sistema de álgebra computacional con las mejores características, manteniendo el código lo más simple posible para que sea comprensible.

`SymPy` ya viene incluido en la instalación estándar de Anaconda por defecto. Para comenzar a usarlo, solo debemos de importarlo:

**Referencia:**
- http://www.sympy.org/en/index.html

````python
import sympy as sym
````

````python
sym.pi
````

````python
sym.sin(sym.pi)
````

````python
np.sin(np.pi)
````

````python
import numpy as np
import sympy as sym
````

````python
# Definimos una variable simbólica
x=sym.var('x')

# Definimos una variable usando 'x' como simbólica
f=sym.sin(x)*sym.exp(-x)
f
````

````python
# Derivamos
sym.diff(f,x) # Con respecto a 'x'
````

````python
sym.diff(f)
````

````python
# Derivando 'f' dos veces con respecto a 'x'
sym.diff(f,x,2)
````

````python
# Para encontrar las raíces
sym.solve(f)
````

````python
# Gráfica de 'f'
sym.plot(f)
````

````python
sym.plot(f,(x,0,2*sym.pi)) # Limitamos el rango
````

### PyPlot de matplotlib

<img style="float: right; margin: 0px 0px 15px 15px;" src="https://matplotlib.org/_static/logo2.svg" width="300px" height="100px" />

El módulo `PyPlot` de la librería `matplotlib` contiene funciones que nos permite generar una gran cantidad de gráficas rápidamente. Las funciones de este módulo están escritas con el mismo nombre que las funciones para graficar en `Matlab`.

**Referencia:**
- https://matplotlib.org/api/pyplot_summary.html

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
x=np.linspace(0,2*np.pi,500) # Genera un arreglo de 500 elementos entre 0 y 2pi
y=np.sin(x)                  # Evalua la función seno elemento a elemento (500 veces pues)
````

````python
plt.plot(x,y) # Grafíca los elementos 'x' vs 'y'
plt.grid()    # Muestra la regilla
````

````python
plt.plot(x,y,'r')   # 'r'  ->  muestra la línea roja
plt.xlabel('x')     # Etiqueta el eje 'x'
plt.ylabel('y')
plt.title('sin(x)') # Título de la figura
plt.grid()
````

````python
# Ayudín
# plt.plot?
````

````python
plt.plot(x,y,':g')                   # Mostramos el seno como una línea punteada verde
puntos=np.array([0,np.pi,2*np.pi])   # Puntos de interés
plt.plot(puntos,np.sin(puntos),'xr') # Mostramos los puntos de interés con cruces rojas
plt.grid()
````

````python
plt.plot(x,y,':g',lw=4)                    # 'lw'='linewidth'
puntos=np.array([0,np.pi,2*np.pi])         # Salu2
plt.plot(puntos,np.sin(puntos),'xr',ms=20) # 'ms'='markersize'
plt.grid()
````

````python
y.shape # Tenemos 500 elementos en 'y'
````

````python
plt.plot(y) # Ordena los valores de 'y' de 0 a 499
````

````python
# Para graficar múltiples líneas, damos simpre pares de elementos 'x' y 'y'
plt.plot(x,np.sin(x),
         x,np.cos(x),
         x,np.exp(-0.5*x)*np.sin(2*x))
plt.legend(['seno','coseno','seno atenuado'],loc='center left')
plt.grid()
````

````python
# plt.legend?
"""
    ==================   =============
    Location String      Location Code
    ==================   =============
    'best' (Axes only)   0
    'upper right'        1
    'upper left'         2
    'lower left'         3
    'lower right'        4
    'right'              5
    'center left'        6
    'center right'       7
    'lower center'       8
    'upper center'       9
    'center'             10
    ==================   =============
"""
````

````python
plt.plot(x,np.sin(x),x,np.cos(x),x,np.exp(-0.5*x)*np.sin(2*x))
# Posicionamos el legend usando coordenadas relativas a la figura
plt.legend(['seno','coseno','seno atenuado'],bbox_to_anchor=(1.01,0.5),loc=6) # En este caso 'loc' es el punto del legend donde posiciona las coordenadas
plt.grid()
````

````python
plt.plot(x,np.sin(x),x,np.cos(x),x,np.tan(x))
plt.ylim(-5,5)               # límite para el eje 'y'
# plt.axis([0,2*np.pi,-5,5]) # límite para ambos ejes (x_min, x_max, y_min, y_max)
plt.grid()
````

# Guía para presentación de proyectos.
> Se especifican los componentes básicos que deberá tener cada proyecto. El proyecto debe ser un problema que se pueda modelar con las herramientas vistas en el módulo. En el caso del tercer módulo: Métodos de búsqueda.

___
## Especificaciones

1. Se asumen los equipos planteados en el proyecto 2.
2. A cada equipo se le asignará un problema a resolver para hacerlo con el tema de Métodos de búsqueda (Enfriamiento simulado).
3. Debe contener
    * Titulo
    * Objetivos
    * Modelo
    * Solución
    * Conclusiones
    * Referencias

### Entregable - Notebook de jupyter.
> Los anteriores numerales los deben desarrollar en un notebook de jupyter y llamarlo de la siguiente manera `ProyectoModulo1_ApellidoN1_ApellidoN2_ApellidoN3`, donde `ApellidoNi` hace referencia al apellido paterno y la inicial del primer nombre del integrante `i`.

### Presentación.
> En este caso, el 100% de la calificación corresponde al entregable del proyecto.

___
## Componentes básicos básicos

### Título del trabajo.
> Debe describir el trabajo.

### Objetivos.
> - Se refieren a los propósitos por los cuales se hace el trabajo.
> - Deben ser concretos, evaluables y verificables.
> - Deben ser escritos en infinitivo.
> #### Objetivo general.
>  - Fin último de estudio. Se formula atendiendo el propósito global del trabajo. No presenta detalles.
>  - Se orienta a la totalidad del trabajo. Tiene que ver con el título del trabajo.
> #### Objetivos específicos
>  - De forma detallada describen cada uno de los componentes del trabajo.
>  - En conjunto garantizan la consecución del objetivo general.

Referencia:
- https://es.slideshare.net/rosbur/metodologia-objetivos-generales-y-especficos

### Modelo que representa el problema.
> - Se debe incluir la función de energía a optimizar.
> - Deducción del modelo. Explicación detallada de las ecuaciones de acuerdo al problema que se quiera resolver o modelo que se quiera representar.
> - ¿Qué situación representa el modelo? ¿Cuáles son las limitaciones fundamentales?
> - Significado y valor de los parámetros (constantes que aparezcan en el modelo).

### Solución del problema de optimización.
> - Se debe encontrar una solución óptima del estado.
> - ¿Es factible la solución?
> - Visualización de la solución del problema.
>
>   Se deben mostrar gráficas y/o tablas que ilustren de forma adecuada los resultados. No olviden etiquetar los ejes y si se tienen varias curvas en una sola gráfica también etiquetarlas (ojo: que las curvas se puedan diferenciar con colores y estilos de línea).

### Conclusiones.
> Mucho cuidado, las conclusiones no son cualquier cosa. Se debe concluir respecto a los objetivos planteados de acuerdo a los resultados obtenidos.

### Referencias.
> Citar (en formato APA) la bibliografía utilizada.

# Consideraciones importantes
El problema relativo a cada propuesta de proyecto se asignarán de forma individual en el apartado correspondiente. A partir del mismo, defina los objetivos, encuentre la solución y redacte las conclusiones.

Si tiene que consultar una IA generativa, la norma APA7 señala que se debe especificar la versión del modelo y la fecha de consulta, idealmente también se debe incluir el *prompt* (pero eso no lo dice la norma).

Considere en su redacción que el mayor peso de la calificación lo llevará la interpretación de los resultados y su enfoque en los objetivos, ya que se trata de un indicador de dominio sobre lo que ocurre en el plano abstracto/matemático que se ve reflejado en el plano real (principal objetivo de la materia).

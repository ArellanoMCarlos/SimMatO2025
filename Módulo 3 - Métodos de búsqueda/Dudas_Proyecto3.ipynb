```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# (Ck_1-C)/dt=f
dt=0.01
dx=0.1
t=0
C=np.zeros(int(10/dx)+1)
C[0]=100
D=0.005
k=0.01
u=0.01
while t<10:
  Cx_1=C.copy()
  Cx_1[2:]=Cx_1[:-2] # C(t,x+dx)
  Cx=C.copy()
  Cx[:-2]=Cx[2:]     # C(t,x-dx)
  f=D*(Cx_1-Cx)/(2*dx)-k*C-u*(Cx_1-2*C+Cx)/(dx**2)

  Ck_1=C+f*dt
  t=t+dt
  C=Ck_1
plt.plot(C)
```
